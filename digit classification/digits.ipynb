```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_train = pd.read_csv('zip.train.txt', header = None,delimiter = " ")
data_test = pd.read_csv('zip.test.txt',header = None,delimiter=" ")

data_train = data_train[data_train.iloc[:, 0].isin([1,4])]
data_test = data_test[data_test.iloc[:, 0].isin([1,4])]
data_train.describe()
```

```python
def symmetry(df):
    img = np.array(df.iloc[1:257])
    img = img.reshape((16,16))
    inv = img[:,::-1]
    diff = (inv - img)
    return (16 * 16 - np.count_nonzero(diff))/(16*16.0)
classify = lambda df: 1 if df.iloc[0]==1 else 0
intensity = lambda df: df.iloc[1:257].sum()/(16*16)

data_train['symmetry'] = data_train.apply(symmetry,axis=1)
data_train['intensity'] = data_train.apply(intensity, axis=1)
data_train['class'] = data_train.apply(classify,axis = 1)
data_train.head()
```

```python
data_test['symmetry'] = data_test.apply(symmetry,axis=1)
data_test['intensity'] = data_test.apply(intensity, axis=1)
data_test['class'] = data_test.apply(classify,axis = 1)
data_test.head()
```

```python
x_train = np.array(data_train_new.loc[:,['intensity','symmetry']])
y_train = np.array(data_train_new["class"])

x_test = np.array(data_test_new.loc[:,['intensity','symmetry']])
y_test = np.array(data_test_new["class"])
```

```python
sigmoid = lambda z: 1/(1+np.exp(-z))
def costFunction(theta, X, y):
#     print(X.shape, y.shape, theta.shape)
    m=len(y)
    predictions = sigmoid(np.dot(X,theta))
    error = (-y * np.log(predictions)) - ((1-y)*np.log(1-predictions))
    cost = 1/m * sum(error)
    grad = 1/m * np.dot(X.transpose(),(predictions - y))
    return cost[0] , grad

m , n = x_train.shape[0], x_train.shape[1]
x_train= np.append(np.ones((m,1)),x_train,axis=1)
y_train=y_train.reshape(m,1)
initial_theta = np.zeros((n+1,1))


m , n = x_test.shape[0], x_test.shape[1]
x_test= np.append(np.ones((m,1)),x_test,axis=1)
y_test=y_test.reshape(m,1)

cost, grad= costFunction(initial_theta,x_train,y_train)
print("Cost of initial theta is",cost)
print("Gradient at initial theta (zeros):",grad)
```

```python
def gradientDescent(X,y,theta,alpha,num_iters):
    m=len(y)
    J_history =[]
    
    for i in range(num_iters):
        cost, grad = costFunction(theta,X,y)
        theta = theta - (alpha * grad)
        J_history.append(cost)
    
    return theta , J_history

theta, J_history = gradientDescent(x_train,y_train,initial_theta,0.1,1000)
print(theta)
```

```python
def computeError(X, y, theta):
    predictions = sigmoid(np.dot(X,theta))
    error = (-y * np.log(predictions)) - ((1-y)*np.log(1-predictions))
    return error
print("Train data Error: ", np.mean(computeError(x_train, y_train, theta)))
print("Test data Error: ", np.mean(computeError(x_test, y_test, theta)))
```

```python
def classifierPredict(theta,X):
    predictions = X.dot(theta)
    return predictions>0
p=classifierPredict(theta,x_train)
print("Train data score:", sum(p==y_train)[0]/len(y_train))
p=classifierPredict(theta,x_test)
print("Train data score:", sum(p==y_test)[0]/len(y_test))
```
